```python
import requests
import pandas
```

```python
offset_start = 0
offset_end = 10
limit = 100
```

```python
url = 'https://xn--80az8a.xn--d1aqf.xn--p1ai/сервисы/api/kn/object?offset={offset}&limit={limit}&sortType=asc'
dataframe_list = []
for i in range(offset_start, offset_end + 1):
    request = requests.get(url.format(offset=i, limit=limit))
    request_json = request.json()
    # check if data empty => there is no more objects
    if not request_json["data"]["list"]:
        break
    dataframe_list.append(pandas.DataFrame(request_json["data"]["list"]))
```

```python
full_result_df = pandas.concat(dataframe_list)
full_result_df.describe()
```

```python
def get_image_url(request):
    if request.status_code == 200:
        req_json = request.json()
        if req_json.get('data', {}).get('photoRenderDTO', {}):
            photo_list_objs = req_json['data']['photoRenderDTO']
            # select only first photo
            return photo_list_objs[0].get('objRenderPhotoUrl', None)
```

```python
# save in current folder
import os
path = os.path.abspath(os.getcwd())
# create folder for images
try:
    os.mkdir('images')
except FileExistsError:
    # if folder already exists
    pass

images_folder = path + '/images'
# get all objects id for url
objects_id = result_df['id'].unique().tolist()
object_url = 'https://xn--80az8a.xn--d1aqf.xn--p1ai/%D1%81%D0%B5%D1%80%D0%B2%D0%B8%D1%81%D1%8B/api/object/{obj_id}'

for object_id in objects_id:
    request = requests.get(object_url.format(obj_id=object_id), stream=True)
    image_url = get_image_url(request)
    if image_url:
        request_image = requests.get(image_url)
        image_name = f'{images_folder}/{object_id}.jpg'
        with open(image_name, 'wb') as f:
            f.write(request_image.content)
```
